```python
import numpy as np
```

```python
# Завдання 1. Створіть одновимірний масив (вектор) з першими 20-ма натуральними числами та виведіть його значення.
task1_array = np.arange(1, 21)
print("Вектор з першими 20-ма натуральними числами:\n", task1_array)
```

```python
# Завдання 2. Створіть двовимірний масив (матрицю) розміром 4x4, заповніть його одиницями та виведіть його значення.
task2_matrix = np.ones((4, 4))
print("Матриця 4x4 з одиницями:\n", task2_matrix)
```

```python
# Завдання 3. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від -1 до 1 та виведіть його значення.
task3_matrix = np.random.uniform(-1, 1, (5, 5))
print("Матриця 5x5 з дійсними числами від -1 до 1:\n", task3_matrix)
```

```python
# Завдання 4. Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та виведіть його значення.
task4_matrix = np.random.rand(3, 3)
print("Матриця 3x3 з дійсними числами від 0 до 1:\n", task4_matrix)
```

```python
# Завдання 5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від 1 до 20 та виконайте на них поелементні операції додавання, віднімання та множення.
array_a = np.random.randint(1, 21, 6)
array_b = np.random.randint(1, 21, 6)

print("Масив A:", array_a)
print("Масив B:", array_b)
print("Додавання (A + B):", array_a + array_b)
print("Віднімання (A - B):", array_a - array_b)
print("Множення (A * B):", array_a * array_b)
```

```python
# Завдання 6. Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.
matrix_3x2 = np.random.randint(1, 6, (3, 2))
matrix_2x4 = np.random.randint(1, 6, (2, 4))
product_matrix = np.dot(matrix_3x2, matrix_2x4)

print("Матриця 3x2:\n", matrix_3x2)
print("Матриця 2x4:\n", matrix_2x4)
print("Результат множення (матриця 3x4):\n", product_matrix)
```

```python
# Завдання 7. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.
task7_matrix = np.random.rand(3, 3)
print("Оригінальна матриця 3x3:\n", task7_matrix)
print("Транспонована матриця:\n", task7_matrix.T)
```

```python
# Завдання 8. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.
mat1_task8 = np.random.randint(1, 11, (3, 3))
mat2_task8 = np.random.randint(1, 11, (3, 3))

print("Матриця 1:\n", mat1_task8)
print("Матриця 2:\n", mat2_task8)
print("Різниця (Матриця 1 - Матриця 2):\n", mat1_task8 - mat2_task8)
```

```python
# Завдання 9. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.
mat1_task9 = np.random.randint(-5, 6, (3, 3))
mat2_task9 = np.random.randint(-5, 6, (3, 3))
product_task9 = np.dot(mat1_task9, mat2_task9)

print("Матриця 1:\n", mat1_task9)
print("Матриця 2:\n", mat2_task9)
print("Добуток матриць:\n", product_task9)
```

```python
# Завдання 10. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
task10_matrix = np.random.uniform(-1, 1, (3, 3))
row_sums_vector = np.sum(task10_matrix, axis=1).reshape(-1, 1)

print("Матриця 3x3:\n", task10_matrix)
print("Вектор-стовпчик із сумами рядків:\n", row_sums_vector)
```
