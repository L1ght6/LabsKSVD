```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завдання 1: Global YouTube Statistics

# 1. Читання CSV (використовуємо кодування windows-1252)
df_yt = pd.read_csv('Global YouTube Statistics.csv', encoding='windows-1252')

# 2. Перші 5 рядків
display(df_yt.head())

# 3. Розмір датасету
print("Розмір датасету:", df_yt.shape)

# 4. Перевірка пропусків
print("Пропуски до обробки:\n", df_yt.isna().sum())

# 5. Заміна пропусків числових стовпців на середнє та зміна типу на float
numeric_cols = df_yt.select_dtypes(include=[np.number]).columns
df_yt[numeric_cols] = df_yt[numeric_cols].fillna(df_yt[numeric_cols].mean()).astype(float)

# 6. Повторна перевірка
print("Пропуски після обробки:\n", df_yt.isna().sum())

# 7. Кількість унікальних країн
print("Унікальних країн:", df_yt['Country'].nunique())

# 8. Діаграма розподілу переглядів
df_yt['video views'].plot(kind='hist', bins=50, title='Distribution of Video Views', figsize=(10,6))
plt.xlabel('Video Views')
plt.show()

# 9. Макс, мін, середнє значення переглядів
print("Максимальні перегляди:", df_yt['video views'].max())
print("Мінімальні перегляди:", df_yt['video views'].min())
print("Середні перегляди:", df_yt['video views'].mean())

# 10. Країна з найбільшою кількістю завантажених відео
country_uploads = df_yt.groupby('Country')['uploads'].sum()
print("Країна з найбільшою кількістю завантажень:", country_uploads.idxmax())

# 11. Назва з найбільшою та найменшою кількістю uploads
print("Найбільше завантажень (Title):", df_yt.loc[df_yt['uploads'].idxmax(), 'Title'])
print("Найменше завантажень (Title):", df_yt.loc[df_yt['uploads'].idxmin(), 'Title'])



# Завдання 2: Amazon Top 50 Bestselling Books

# 1. Читання CSV
df_books = pd.read_csv('bestsellers with categories.csv')

# Виведення перших 10 рядків та розміру
display(df_books.head(10))
print("Розмір датасету:", df_books.shape)
print("Датасет зберігає дані про", df_books.shape[0], "книг.")

# Перейменування колонок
df_books.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

# Перевірка пропусків
print("Пропуски по стовпцях:\n", df_books.isna().sum())
print("Чи є пропуски взагалі?", df_books.isna().values.any()) # Виведе False (Ні)

# Унікальні жанри
print("Унікальні жанри:", df_books['genre'].unique())

# Максимальна, мінімальна, середня та медіанна ціна
print("Максимальна ціна:", df_books['price'].max())
print("Мінімальна ціна:", df_books['price'].min())
print("Середня ціна:", df_books['price'].mean())
print("Медіанна ціна:", df_books['price'].median())

# Завдання з пошуку
max_rating = df_books['user_rating'].max()
print("Найвищий рейтинг:", max_rating)
print("Кількість книг з найвищим рейтингом:", len(df_books[df_books['user_rating'] == max_rating]))

print("Книга з найбільшою кількістю відгуків:", df_books.loc[df_books['reviews'].idxmax(), 'name'])

books_2010 = df_books[df_books['year'] == 2010]
print("Найдорожча книга 2010 року:", books_2010.loc[books_2010['price'].idxmax(), 'name'])

fiction_2012 = df_books[(df_books['genre'] == 'Fiction') & (df_books['year'] == 2012)]
print("Кількість книг Fiction у 2012 році:", len(fiction_2012))

books_4_9 = df_books[(df_books['user_rating'] == 4.9) & (df_books['year'].isin([2010, 2011]))]
print("Книг з рейтингом 4.9 у 2010 та 2011 роках:", len(books_4_9))

# Групування за жанром (макс та мін ціна)
print("Ціна за жанрами:\n", df_books.groupby('genre').agg({'price': ['max', 'min']}))
```
