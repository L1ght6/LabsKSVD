Імпорт бібліотек та завантаження даних

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Завантаження датасетів
iris = px.data.iris()
tips = px.data.tips()
gapminder = px.data.gapminder()
```

Завдання 1 - Точкові графіки для датасету Iris

```python
# 1. Точковий графік з дискретними кольорами
fig1 = px.scatter(iris, x="sepal_length", y="sepal_width", size="petal_length",
                  color="species",
                  title="Iris: Дискретні кольори")
fig1.update_layout(title_x=0.5)
fig1.show()
```

```python
# 2. Точковий графік з неперервними кольорами (використовуємо species_id як числове значення)
fig2 = px.scatter(iris, x="sepal_length", y="sepal_width", size="petal_length",
                  color="species_id", color_continuous_scale=px.colors.sequential.Plasma,
                  title="Iris: Неперервні кольори")
fig2.update_layout(title_x=0.5)
fig2.show()
```

```python
# 3. Точковий графік з різними маркерами
fig3 = px.scatter(iris, x="sepal_length", y="sepal_width",
                  color="species", symbol="species",
                  title="Iris: Різні маркери")
fig3.update_layout(title_x=0.5)
fig3.show()
```

Завдання 2

```python
# Завдання 2: Лінія тренду для датасету tips
fig_tips = px.scatter(tips, x="total_bill", y="tip", trendline="ols",
                      title="Залежність чайових від суми рахунку з лінією тренду")
fig_tips.update_layout(title_x=0.5)
fig_tips.show()
```

Завдання 3 та 4 - Датасет Gapminder

```python
# Завдання 3: Взаємозв'язок між ВВП (доходом) та тривалістю життя
fig_gap1 = px.scatter(gapminder, x="gdpPercap", y="lifeExp",
                      animation_frame="year", animation_group="country",
                      size="pop", color="continent", hover_name="country",
                      log_x=True, size_max=55,
                      title="Взаємозв'язок: Рівень доходу та Тривалість життя")
fig_gap1.update_layout(title_x=0.5)
fig_gap1.show()

```

```python
# Завдання 4: Динаміка населення Японії за роками
japan_data = gapminder[gapminder['country'] == 'Japan']
fig_gap2 = px.line(japan_data, x="year", y="pop", markers=True,
                   title="Динаміка населення Японії за роками")
fig_gap2.update_layout(title_x=0.5)
fig_gap2.show()
```

```python
# Завдання 5: Box Plot для розподілу довжини пелюстки (petal_length)
fig_box = px.box(iris, x="species", y="petal_length", color="species",
                 title="Розподіл довжини пелюстки для різних видів ірису")
fig_box.update_layout(title_x=0.5)
fig_box.show()
```

Завдання 6 - Тригонометричні функції

```python
# 1. Функція y = 2cos(x - π/4) на проміжку [0; 5π]
x1 = np.linspace(0, 5*np.pi, 500)
y1 = 2 * np.cos(x1 - np.pi/4)

fig_trig1 = go.Figure(data=go.Scatter(x=x1, y=y1, mode='lines'))
fig_trig1.update_layout(title='Графік функції $y = 2\cos(x - \frac{\pi}{4})$', title_x=0.5)
fig_trig1.show()
```

```python
# 2. Функція y = 1/2 sin(2x + π/3) на проміжку [0; 2π]
x2 = np.linspace(0, 2*np.pi, 500)
y2 = 0.5 * np.sin(2*x2 + np.pi/3)

fig_trig2 = go.Figure(data=go.Scatter(x=x2, y=y2, mode='lines', line=dict(color='green')))
fig_trig2.update_layout(title='Графік функції $y = \frac{1}{2}\sin(2x + \frac{\pi}{3})$', title_x=0.5)
fig_trig2.show()
```

```python
# 3. Функція y = 2tg(1/2x + π/6) на проміжку [-2π; 2π]
x3 = np.linspace(-2*np.pi, 2*np.pi, 1000)
y3 = 2 * np.tan(0.5*x3 + np.pi/6)

# Прибираємо "вертикальні лінії" (асимптоти), де тангенс прямує до нескінченності
y3[:-1][np.diff(y3) < 0] = np.nan

fig_trig3 = go.Figure(data=go.Scatter(x=x3, y=y3, mode='lines', line=dict(color='red')))
fig_trig3.update_layout(title='Графік функції $y = 2\tan(\frac{1}{2}x + \frac{\pi}{6})$',
                        title_x=0.5, yaxis_range=[-10, 10])
fig_trig3.show()
```

Імпорт та підготовка даних

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Завантаження датасетів
iris = px.data.iris()
gapminder = px.data.gapminder()
```

Датасет Iris (3D, Гістограма та Середні значення)

```python
# 1. 3D-графік розподілу
fig_3d = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_length',
                        color='species',
                        title='3D Розподіл характеристик Iris')
fig_3d.update_layout(title_x=0.5)
fig_3d.show()
```

```python

# 2. Гістограма розподілу "petal_width"
fig_hist = px.histogram(iris, x="petal_width", color="species",
                         marginal="rug", # Додає позначки щільності знизу
                         title='Розподіл ширини пелюстки (petal_width) за видами')
fig_hist.update_layout(title_x=0.5, barmode='overlay')
fig_hist.show()
```

```python
# 3. Лінійний графік середнього sepal_length в залежності від sepal_width
iris_grouped = iris.groupby(['species', 'sepal_width'])['sepal_length'].mean().reset_index()
fig_mean_line = px.line(iris_grouped, x="sepal_width", y="sepal_length", color="species",
                        title="Зміна середнього sepal_length від sepal_width")
fig_mean_line.update_layout(title_x=0.5)
fig_mean_line.show()
```

Датасет Gapminder (Економіка та Демографія)

```python
# Обчислення загального ВВП
gapminder['total_gdp'] = gapminder['gdpPercap'] * gapminder['pop']

# 1. Гістограма (Bar chart) 10 країн з найбільшим ВВП (2007)
gap2007_top10 = gapminder[gapminder['year'] == 2007].nlargest(10, 'total_gdp')
fig_gdp_bar = px.bar(gap2007_top10, x='country', y='total_gdp', color='total_gdp',
                     text='total_gdp', color_continuous_scale='Viridis',
                     title='Топ-10 країн за рівнем ВВП (2007)')
fig_gdp_bar.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig_gdp_bar.update_layout(title_x=0.5)
fig_gdp_bar.show()
```

```python
# 2. Кругова діаграма ВВП Європи (Топ-5)
eu_top5 = gapminder[(gapminder['continent'] == 'Europe') & (gapminder['year'] == 2007)].nlargest(5, 'total_gdp')
fig_pie = px.pie(eu_top5, values='total_gdp', names='country',
                 title='Розподіл ВВП серед топ-5 країн Європи')
fig_pie.update_layout(title_x=0.5)
fig_pie.show()
```

```python
# 3. Стовпчаста діаграма динаміки населення (Канада та Океанія)
# Оскільки Канада в Північній Америці, ми об'єднаємо її з країнами Океанії для порівняння
oceania_canada = gapminder[(gapminder['continent'] == 'Oceania') | (gapminder['country'] == 'Canada')]
fig_pop = px.bar(oceania_canada, x='year', y='pop', color='country', barmode='group',
                 title='Динаміка зміни кількості населення (Канада та Океанія)')
fig_pop.update_layout(title_x=0.5)
fig_pop.show()
```

Тригонометричні функції

```python
# 1. y = 2cos(x - π/4) на [0; 5π]
x1 = np.linspace(0, 5 * np.pi, 500)
y1 = 2 * np.cos(x1 - np.pi/4)
fig_t1 = go.Figure(data=go.Scatter(x=x1, y=y1, mode='lines'))
fig_t1.update_layout(title='Графік y = 2cos(x - π/4)', title_x=0.5)
fig_t1.show()
```

```python
# 2. y = 1/2 sin(2x + π/3) на [0; 2π]
x2 = np.linspace(0, 2 * np.pi, 500)
y2 = 0.5 * np.sin(2*x2 + np.pi/3)
fig_t2 = go.Figure(data=go.Scatter(x=x2, y=y2, mode='lines', line=dict(color='orange')))
fig_t2.update_layout(title='Графік y = 1/2 sin(2x + π/3)', title_x=0.5)
fig_t2.show()
```

```python
# 3. y = 2tg(1/2x + π/6) на [-2π; 2π]
x3 = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y3 = 2 * np.tan(0.5*x3 + np.pi/6)

# Маскування розривів тангенса для коректного відображення
y3[:-1][np.diff(y3) < 0] = np.nan

fig_t3 = go.Figure(data=go.Scatter(x=x3, y=y3, mode='lines', line=dict(color='purple')))
fig_t3.update_layout(title='Графік y = 2tg(1/2x + π/6)', title_x=0.5, yaxis_range=[-15, 15])
fig_t3.show()
```
